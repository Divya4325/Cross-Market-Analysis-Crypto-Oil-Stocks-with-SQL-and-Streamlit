```python
import requests
import pandas as pd
import time
all_data=[]
base_url="https://api.coingecko.com/api/v3/coins/markets"
for page in range(1,6):
  params={
      "vs_currency":"inr",
      "per_page":250,
      "order":"market_cap_desc",
      "page":page,
      "sparkline":"false"
  }
  while True:
     response=requests.get(base_url,params=params)

     if response.status_code==429:
       print("Rate limit hit.Waiting 10 seconds")
       time.sleep(10)
       continue

     if response.status_code!=200:
       print("Error:", response.text)
       break
     data=response.json()
     print(f"Page{page}---->{len(data)}coins")

     all_data.extend(data)
     break

     time.sleep(3)
  print("Total coins collected:",len(all_data))
```

```python
import requests
import pandas as pd
import time

all_prices=[]

for coin in coin_ids:
    url = f"https://api.coingecko.com/api/v3/coins/{coin}/market_chart?vs_currency=inr&days=365"

    response = requests.get(url)
    if response.status_code != 200:
        print(f"Error downloading {coin}: {response.status_code}")
        continue

    data = response.json()
    if 'prices' not in data :
        print(f"No price data for {coin}")
        print(data)
        continue

    prices = data['prices']
    for p in prices:
      timestamp=p[0]
      price=p[1]
      date=pd.to_datetime(timestamp,unit='ms').date()
      all_prices.append([coin,date,price])

    print(f"Downloaded:{coin}")
    time.sleep(35)
```

```python
import yfinance as yf
import pandas as pd
tickers = ["^GSPC", "^IXIC", "^NSEI"]
start_date = "2020-01-01"
end_date = "2025-09-30"
all_stocks=[]
for ticker_name in tickers: # Renamed loop variable to avoid conflict with `df["Ticker"]`
    df_temp = yf.download(ticker_name, start=start_date, end=end_date)
    if df_temp.empty:
        print(f"No Data for {ticker_name}")
        continue
    df_temp = df_temp.reset_index()
    # Select and rename columns to match the database schema
    df_selected = df_temp[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']].copy()
    df_selected.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
    df_selected['ticker'] = ticker_name
    all_stocks.append(df_selected)

final_df=pd.concat(all_stocks,ignore_index=True)
final_df
```

**5.Get the % change in Bitcoin’s price between Sep 2024 and Sep 2025**

```python
cursor.execute("""
    SELECT ((price_inr_sep2025-price_inr_sep2024)*100.0/
       price_inr_sep2025) AS percent_change
     FROM (
       SELECT
         (SELECT price_inr
        FROM crypto_prices
        WHERE coin_id='bitcoin'
        AND date BETWEEN '2025-09-01'AND '2025-09-30'
        ORDER BY date DESC LIMIT 1) AS price_inr_sep2025,

        (SELECT price_inr
        FROM crypto_prices
        WHERE coin_id='bitcoin'
        AND date BETWEEN '2024-09-01'AND '2024-09-30'
        ORDER BY date DESC LIMIT 1) AS price_inr_sep2024
) AS subquery_prices;
""")
result=cursor.fetchone()
print("Percentage change in Bitcoin's price between Sep 2024 and Sep 2025:",result[0])
```

** Join Queries (Cross-Market Analysis)**

*1.Compare Bitcoin vs Oil average price in 2025*


```python
cursor.execute("""
    SELECT
      ROUND(AVG(c.price_inr),2) AS bitcoin_avg_2025,
      ROUND(AVG(o.price_inr),2) AS oil_avg_2025
     FROM crypto_prices c
     JOIN oil_prices o
       ON c.date=o.date
     WHERE c.coin_id='bitcoin'
     AND strftime('%Y',c.date)='2025';
""")
result=cursor.fetchone()
print("Bitcoin Avg Price in 2025:",result[0])
print("Oil Avg Price in 2025:",result[0])
```

**2.Check if Bitcoin moves with S&P 500 (correlation idea)**

```python
cursor.execute("""
    SELECT
      c.date,
      c.price_inr AS bitcoin_price,
      s.close AS sp500_price
    FROM crypto_prices c
    JOIN stock_prices s
       ON c.date=s.date
    WHERE c.coin_id='bitcoin'
    AND s.ticker='^GSPC'
    ORDER BY c.date;
""")
rows = cursor.fetchall()
correlation_df = pd.DataFrame(rows, columns=['date', 'bitcoin_price', 'sp500_price'])
correlation = correlation_df["bitcoin_price"].corr(correlation_df["sp500_price"])
print("Correlation between Bitcoin and S&P500:", correlation)
```

**3.Compare Ethereum and NASDAQ daily prices for 2025**

**4.Find days when oil price spiked and compare with Bitcoin price change**

```python
cursor.execute("""
  WITH oil_changes AS(
    SELECT date, price_inr,
     ROUND((price_inr-LAG(price_inr) OVER (ORDER BY date))*100.0/LAG(price_inr) OVER (ORDER BY date),2)
      AS oil_pct_change
    FROM oil_prices),
  btc_changes AS(
      SELECT date, price_inr,
       ROUND((price_inr-LAG(price_inr) OVER (ORDER BY date))*100.0/LAG(price_inr) OVER (ORDER BY date),2)
       AS btc_pct_change
     FROM crypto_prices
     WHERE coin_id='bitcoin'
  )
  SELECT
   o.date,
   o.oil_pct_change,
   b.btc_pct_change
 FROM oil_changes o
 JOIN btc_changes b
  ON o.date=b.date
 WHERE o.oil_pct_change >3
 ORDER BY o.date
""")
rows=cursor.fetchall()
for row in rows:
  print("Date:",row[0])
  print("Oil % Change:",row[1])
  print("Bitcoin % Change:",row[2])
```
